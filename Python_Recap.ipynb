#OOP classes inheritance and whatever

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def speak(self):
    print("hello")

class Dog(Animal):
  def __init__(self, name, age):
    super().__init__(name, age)

  def speak(self):
    print("woof")

x = Animal("dog", 10).speak()
x = Dog("dog", 10).speak()
```

#Calculator App

```python
#class based calc
#basic operatorrs
#loop for repeated use
#history stored in list
#save and load history using text file
#simple error hand
```

```python
HISTORY_FILE = "new.txt"
```

```python
class Calculator:
  def __init__(self):
    self.history = list()

  def add(self, num1, num2):
    result  = num1+num2
    print(result)
    self.history.append(f"{num1}+{num2}={result}")
    self.save_history()

  def subtract(self, num1,num2):
    result = num1-num2
    print(result)
    self.history.append(f"{num1}-{num2}={result}")
    self.save_history()

  def save_history(self):
    with open(HISTORY_FILE,"w") as file:
      for item in self.history:
        file.write(f"{item}\n")

  def load_history(self):
    with open(HISTORY_FILE,"r") as file:
      self.history = file.readlines()
      for item in self.history:
        if(item == "\n"):
          continue
        print(item.strip())
```

```python
calculator = Calculator()
```

```python
def ask_input():
  try:
    num1 = int(input("enter first number: "))
    num2 = int(input("enter second number: "))
    return num1,num2
  except ValueError:
    print("Invalid Input")
    return False,False
```

```python
while True:
  user = float(input("Select an option:\n1.Add\n2.Subtract\n5.Load History\n6.Exit\n"))
  match user:
    case 1:
      num1,num2 = ask_input()
      if num1 and num2:
        calculator.add(num1,num2)
    case 2:
      num1,num2 = ask_input()
      if num1 and num2:
        calculator.subtract(num1,num2)
    case 5:
      calculator.load_history()
    case 6:
      break
    case _:
      pass
```
