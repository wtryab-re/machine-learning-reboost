```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

Import Dataset

```python
original_data = load_iris().data
original_target = load_iris().target
original_target_names = load_iris().target_names
original_columns = load_iris().feature_names
original_desc = load_iris().DESCR

##shorter version to do this is
#load_iris(as_frame=True).frame
```

Conversions and Combination of target w the features

```python
df = pd.DataFrame(original_data, columns=original_columns)
df["target"]=original_target
df.head()
```

```python
##adding in the target_names
target_names = dict(enumerate(original_target_names))
target_names
```

```python
df["target_name"] = df["target"].map(target_names)
```

Pair Plot

```python
plt.figure(figsize=(6,6))
sns.set_style("darkgrid")
graph_df = df.drop(columns=["target"])
sns.pairplot(graph_df, hue="target_name", diag_kind="kde")
```

#Heatmap

```python
correlation_matrix = graph_df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, linewidths=1, annot=True, cmap="coolwarm")
plt.show()
## the reason why I do not make a heatmap with the target data is because TARGET DATA is categorical in this case.
#with higher correlation we can do pca, and remove some input features.
```

countplot

```python
plt.figure(figsize=(4,4))
sns.countplot(data=df, x="target_name")
plt.tight_layout()
plt.show()
```

scatterplot

```python
#scatterplot compares two continuous values
plt.figure(figsize=(4,4))
sns.scatterplot(data=df, x="petal length (cm)", y="sepal width (cm)", hue="target_name")
plt.title("Sepal Width vs Petal Length")
plt.show()
```

The reason why we create plots is to find relationships, correlations, distributions etc in data.
If we get correlations we can do PCA and remove some input features and
if we can see a clear trends in the output and features we can use it to select models.
